```python
from docx import Document
from docx.shared import Inches, Pt, RGBColor
from docx.enum.text import WD_ALIGN_PARAGRAPH, WD_PARAGRAPH_ALIGNMENT
from docx.enum.table import WD_TABLE_ALIGNMENT
from docx.shared import Cm
# Create a new Document
doc = Document()

# Add heading image
heading_img_path = "/content/drive/MyDrive/img/image10.png"
doc.add_picture(heading_img_path, width=Inches(2))

# Add profile photo
profile_photo_path = "/content/drive/MyDrive/img/image4.jpg"
profile_picture = doc.add_picture(profile_photo_path, width=Inches(2))
profile_picture.alignment = WD_ALIGN_PARAGRAPH.LEFT

def format_table(table):

    # Format paragraphs in cells of the third column
    for cell in table.columns[2].cells:
        for paragraph in cell.paragraphs:
            paragraph.alignment = WD_ALIGN_PARAGRAPH.LEFT

    # Format text in all cells of the table
    for row in table.rows:
        for cell in row.cells:
            for paragraph in cell.paragraphs:
                for run in paragraph.runs:
                    run.font.name = 'Calibri'
                    run.font.size = Pt(12)
                    if run.text == ":":
                        run.font.color.rgb = RGBColor(255, 0, 0)  # Set font color to red for ":"


# Example usage:
# Assuming `document` is your Document object containing tables
for table in doc.tables:
    format_table(table)

# Add text beside the profile picture
profile_text_paragraph = doc.add_paragraph()
run = profile_text_paragraph.add_run("Mahesh Kanchan\n")
run.font.size = Pt(12)
run.font.name = 'Calibri'
run.font.color.rgb = RGBColor(128, 0, 128)
run.font.bold = True

# Add remaining text
runs=profile_text_paragraph.add_run("Vice President Marketing,\n"
                               "Carlsberg India\n\n"
                               "Mahesh possesses over two decades of experience across the FMCG industry. Currently, he serves as the Vice President (Head) of Marketing at Carlsberg India, where he holds the complete Marketing, P&L and On-Trade Sales responsibility for a Gross sales of INR 3735 Crore and marketing investment of 150 Crore. He has been instrumental in making Carlsberg India the fastest growing alco-bev company in India basis a focused power brands (Tuborg & Carlsberg) strategy. Tuborg became the No. 2 brand in 2016, Carlsberg moved from No. 11 (2014) to No. 6 rank in 2016, and No. 5 in 2017.")
runs.alignment = WD_ALIGN_PARAGRAPH.LEFT
runs.font.name = 'Calibri'

# Add a new page
doc.add_page_break()

# Add profile photo for the second page
profile_photo_path = "/content/drive/MyDrive/img/image12.png"
profile_picture = doc.add_picture(profile_photo_path, width=Inches(2))
profile_picture.alignment = WD_ALIGN_PARAGRAPH.LEFT

# Add the "Personal Data" header as a paragraph
personal_data_header = doc.add_paragraph("Personal Data:")
personal_data_header.runs[0].font.color.rgb = RGBColor(128, 0, 128)
personal_data_header.runs[0].font.bold = True
personal_data_header.runs[0].font.name = 'Calibri'
personal_data_header.runs[0].font.size = Pt(12)

# Add a table with three columns
table = doc.add_table(rows=0, cols=3)

# Function to add a row to the table
def add_row(table, left_text, middle_text, right_text):
    row = table.add_row().cells
    row[0].text = left_text
    row[1].text = middle_text
    row[2].text = right_text

# Add rows for personal data
add_row(table, "Date Of Birth", ":", "20/06/2002")
add_row(table, "Nationality", ":", "India")
add_row(table, "Location", ":", "Delhi NCR")
add_row(table, "Languages", ":", "English and Hindi")

for cell in table.columns[2].cells:
    cell.width = Cm(11.0)


# Add Educational Data
# Add the "Educational Data" header as a paragraph
educational_data_header = doc.add_paragraph("Educational Data:")
educational_data_header.runs[0].font.color.rgb = RGBColor(128, 0, 128)
educational_data_header.runs[0].font.bold = True
educational_data_header.runs[0].font.name = 'Calibri'
educational_data_header.runs[0].font.size = Pt(12)

# Add a table for educational data with three columns
table = doc.add_table(rows=0, cols=3)

# Function to add a row to the table
def add_educational_row(table, year, institution, degree):
    row = table.add_row().cells
    row[0].text = year
    row[1].text = ":"
    institution_cell = row[2].paragraphs[0]
    institution_run = institution_cell.add_run(f"{institution}")
    institution_run.font.color.rgb = RGBColor(128, 0, 128)  # Set font color to purple for institution text
    institution_cell.add_run(f"\n{degree}")

# Add rows for educational data
add_educational_row(table, "1997", "Jamnalal Bajaj Institute of Management Studies, Mumbai", "Masters in Management Studies (Marketing)")
add_educational_row(table, "1992", "Narsee Monjee College of Commerce, Mumbai", "Bachelor of Commerce (Financial Accounting)")

for cell in table.columns[2].cells:
    cell.width = Cm(11.0)




# Add image
company_profile_img_path = "/content/drive/MyDrive/img/image0.png"
section = doc.sections[0]
doc.add_picture(company_profile_img_path, width=Inches(7))

# Add "Company Profile:" text
company_profile_paragraph = doc.add_paragraph()
run = company_profile_paragraph.add_run("Company Profile:")
run.font.size = Pt(11)
run.font.name = 'Calibri'
run.font.italic = True
run.font.color.rgb = RGBColor(128, 0, 128)
run.font.bold = True
run1 = company_profile_paragraph.add_run("\tXYZ Solutions Inc.,  Information Technology Services, Gurgaon, India")
run1.font.size = Pt(10)
run1.font.name = 'Calibri'
run1.font.italic = True

key_resp_para = doc.add_heading('Key Responsibilities:', level=1)
key_resp_para.runs[0].font.size = Pt(12)
key_resp_para.runs[0].font.name = 'Calibri'
key_resp_para.runs[0].font.italic = True
key_resp_para.runs[0].font.color.rgb = RGBColor(128, 0, 128)
key_resp_para.runs[0].font.bold = True

# Add bullet points for Key Responsibilities
key_responsibilities = [
    "Responsibility 1",
    "Responsibility 2"
]

for responsibility in key_responsibilities:
    doc.add_paragraph(responsibility, style='ListBullet')

key_ach_para = doc.add_heading('Key Achievements:', level=1)
key_ach_para.runs[0].font.size = Pt(12)
key_ach_para.runs[0].font.name = 'Calibri'
key_ach_para.runs[0].font.italic = True
key_ach_para.runs[0].font.color.rgb = RGBColor(128, 0, 128)
key_ach_para.runs[0].font.bold = True
key_achievements = [
    ("Achievement 1", ["Sub-point 1.1", "Sub-point 1.2"]),
    ("Achievement 2")
]
for achievement in key_achievements:
    if isinstance(achievement, tuple):
        # If achievement is a tuple, it has sub-points
        doc.add_paragraph(achievement[0], style='ListBullet')
        for sub_point in achievement[1]:
            doc.add_paragraph(sub_point, style='ListBullet2')
    else:
        # If achievement is just a string, add it directly as a bullet point
        doc.add_paragraph(achievement, style='ListBullet')

# Add image
company_profile_img_path = "/content/drive/MyDrive/img/image1.png"
section = doc.sections[0]
doc.add_picture(company_profile_img_path, width=Inches(7))

doc.add_page_break()

profile_photo_path = "/content/drive/MyDrive/img/image3.png"
profile_picture = doc.add_picture(profile_photo_path, width=Inches(2))
profile_picture.alignment = WD_ALIGN_PARAGRAPH.LEFT

# Add a table with three columns
table = doc.add_table(rows=0, cols=3)

# Function to add a row to the table
def add_row(table, left_text, middle_text, right_text=None, course=None, info=None):
    row = table.add_row().cells
    row[0].text = left_text
    row[1].text = middle_text
    if right_text is not None:
        row[2].text = right_text

    if right_text is None:
        paragraph = row[2].add_paragraph()
        if course:
            run_course = paragraph.add_run(course)
            run_course.font.color.rgb = RGBColor(128, 0, 128)
        if info:
            paragraph.add_run(f"\n{info}")

# Add rows for additional data with bullet points
add_row(table, "Location Preference", ":", "Delhi NCR", None, None)
add_row(table, "CTC*", ":", "INR x.x Crore fixed + INR xx Lakhs variable.", None, None)
add_row(table, "Contact Details", ":", "+91 87435 94384, fkgjfk.dfsd@gmail.com", None, None)
add_row(table, "2019", ":", None, "London Business School", "Leading for Results Program\nA Leadership Development Program focused on\n •  Strengthening market positions & accelerating growths\n •  Excelling in execution and creating a winning culture")
add_row(table, "1992-1994", ":", None, "Indo-German Training Centre, Mumbai", "Diploma in Business Administration\n •  Sponsored by Bayer (India) Ltd, with 2/3rds of training being on the job.\n •  Course under the aegis of Indo-German Chamber of Commerce.\n•  Training content developed by Deutscher Industrie – und Handelsta (Association of German Chambers of Industry & Commerce).")


for cell in table.columns[2].cells:
    cell.width = Cm(11.0)



# Add a new page
doc.add_page_break()

background_image_path = "/content/drive/MyDrive/img/image8.png"
section = doc.sections[-1]
section.left_margin = Inches(1)
section.right_margin = Inches(1)
section.top_margin = Inches(1)
section.bottom_margin = Inches(0)
section.header_distance = Inches(0)
section.footer_distance = Inches(0)
doc.add_picture(background_image_path, width=Inches(7))


# Add footer image
footer_img_path = "/content/drive/MyDrive/img/image19.jpg"
section = doc.sections[0]
footer = section.footer
footer_paragraph = footer.paragraphs[0]
footer_paragraph.alignment = WD_ALIGN_PARAGRAPH.CENTER
run = footer_paragraph.add_run()
run.add_picture(footer_img_path, width=Inches(7))

# Save the document
doc.save("document_with_images.docx")
```
